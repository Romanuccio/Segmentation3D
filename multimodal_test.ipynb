```python
import torch
from modules.segmentation_model import MultimodalUnet
from models.lightning_models_clifford import Unet3D
```

```python
patch_size = (128, 128, 128)
stride = (64, 64, 64)
device = torch.device(0)
model = Unet3D(
    model=MultimodalUnet,
    depths=[3, 3, 3, 9, 3],
    channel_multipliers=[1, 2, 4, 8, 16],
    modalities = 2,
    embed_dim=64,
    patch_size=patch_size,
    strides=stride,
    padding="same",
    classes=1,
    initial_LR=1e-4,
    loss=None,
    monai=True,
    # metrics=[
    #     ("thresholded_dice", tresholded_dice),
    #     ("tresholded_haus", tresholded_haus),
    # ],
    beta=1,
    # final_activation=torch.nn.Sigmoid(),
    # optimizer=AdamW
).to(device)
```

```python
model.eval()
input = torch.rand([1, 2, 10, 10, 10]).to(device)
with torch.no_grad():
    model.predict_step(input, patch_size=patch_size, strides=stride, padding="same", unpad=True, verbose=False)
```
